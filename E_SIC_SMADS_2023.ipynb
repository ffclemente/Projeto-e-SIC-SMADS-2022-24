```python
# GRÁFICO DE SOLICITAÇÕES DE INFORMAÇÃO (E-SIC) POR MÊS ENTRE 2022 E 2024.

# Carregar pacotes
import matplotlib.pyplot as plt

# Converta a coluna 'Data Recebimento' para o formato de data, lidando com possíveis erros
dados['Data Recebimento'] = pd.to_datetime(dados['Data Recebimento'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Remova o horário (definindo todas as horas para 00:00:00)
dados['Data Recebimento'] = dados['Data Recebimento'].dt.floor('D')

# Remova as linhas com datas inválidas (NaT)
dados = dados.dropna(subset=['Data Recebimento'])

# Agrupe por mês e conte o número de solicitações em cada mês
solicitacoes_por_mes = dados.groupby(dados['Data Recebimento'].dt.to_period("M")).size()

# Cálculo da média mensal
media_mensal = solicitacoes_por_mes.mean()

# Mapeie os períodos para o formato desejado (por exemplo, set-2022)
solicitacoes_por_mes.index = solicitacoes_por_mes.index.strftime('%b / %Y')

# Crie um gráfico de barras com as legendas formatadas e incluindo os valores
plt.figure(figsize=(10, 6))
ax = solicitacoes_por_mes.plot(kind='bar', color='skyblue')

# Adicionar rótulos nas barras
for i, v in enumerate(solicitacoes_por_mes):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')

# Adicionar a média ao gráfico
plt.axhline(y=media_mensal, color='red', linestyle='--', label=f'Média Mensal: {media_mensal:.2f}')
plt.legend()

# Adicionar o valor da média no gráfico
plt.text(len(solicitacoes_por_mes) - 0.5, media_mensal + 0.5, f'Média Mensal: {media_mensal:.2f}', color='red', ha='center', va='bottom')

plt.title('Número de Pedidos de Informação (e-SIC) por Mês SMADS (2022, 2023 e 2024)')
plt.xlabel('Mês')
plt.ylabel('Número de Solicitações')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.show()

print("\nTotal de registros:", len(dados['Protocolo']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dados do arquivo Excel
dados = pd.read_excel('Acompanhamento - Transparência Passiva.xlsx')

# Convertendo a coluna 'Data Recebimento' para o formato de data
dados['Data Recebimento'] = pd.to_datetime(dados['Data Recebimento'], errors='coerce')

# Convertendo a coluna 'Tempo Resposta' para o formato de número (caso esteja em formato de texto)
dados['Tempo Resposta'] = pd.to_numeric(dados['Tempo Resposta'], errors='coerce')

# Removendo linhas com valores nulos em 'Data Recebimento' e 'Tempo Resposta'
dados = dados.dropna(subset=['Data Recebimento', 'Tempo Resposta'])

# Agrupando por ano e calculando a média do tempo de resposta
media_tempo_resposta_por_ano = dados.groupby(dados['Data Recebimento'].dt.to_period("Y"))['Tempo Resposta'].mean()

# Criando um gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(media_tempo_resposta_por_ano.index.astype(str), media_tempo_resposta_por_ano, color='skyblue')

# Adicionando rótulos com os valores no topo das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Personalizando as legendas
legendas_personalizadas = {
    '2022': 'sep. out e nov. 2022',
    '2023': 'jan. a dez. 2023',
    '2024': 'jan., fev. e mar. 2024'
}

# Substituindo as legendas no gráfico
plt.xticks([str(year) for year in media_tempo_resposta_por_ano.index], [legendas_personalizadas.get(str(year), str(year)) for year in media_tempo_resposta_por_ano.index])

plt.title('Tempo de Resposta Médio por Ano')
plt.xlabel('Ano')
plt.ylabel('Tempo de Resposta Médio (horas)')

plt.show()
```

```python
# Carregar pacotes
import matplotlib.pyplot as plt
import numpy as np

# Supondo que você já tenha carregado e pré-processado os dados

# Agrupe por mês e conte o número de solicitações em cada mês
solicitacoes_por_mes = dados.groupby(dados['Data Recebimento'].dt.to_period("M")).size()

# Agrupe por mês e calcule a média do tempo de resposta
media_tempo_resposta_por_mes = dados.groupby(dados['Data Recebimento'].dt.to_period("M"))['Tempo Resposta'].mean()

# Mapeie os períodos para o formato desejado (por exemplo, set-2022)
solicitacoes_por_mes.index = solicitacoes_por_mes.index.strftime('%b / %Y')

# Criar um gráfico de barras lado a lado
fig, ax1 = plt.subplots(figsize=(12, 6))

bar_width = 0.35
index = np.arange(len(solicitacoes_por_mes))

# Barra para o número de solicitações
bars1 = ax1.bar(index, solicitacoes_por_mes, bar_width, color='skyblue', label='Número de Solicitações')

# Adicionar rótulos nas barras de solicitações
for i, v in enumerate(solicitacoes_por_mes):
    ax1.text(i, v + 0.1, str(v), ha='center', va='bottom')

# Adicionar a média ao gráfico (eixo primário)
ax1.legend(loc='upper left')

# Criar eixo secundário (ax2) para o tempo de resposta
ax2 = ax1.twinx()
bars2 = ax2.bar(index + bar_width, media_tempo_resposta_por_mes, bar_width, color='green', alpha=0.5, label='Tempo Médio de Resposta')

# Adicionar a média ao gráfico (eixo secundário)
ax2.axhline(y=media_tempo_resposta_por_mes.mean(), color='green', linestyle='--', label=f'Média Tempo de Resposta: {media_tempo_resposta_por_mes.mean():.2f}')

# Adicionar rótulos nas barras de tempo de resposta
for i, v in enumerate(media_tempo_resposta_por_mes):
    ax2.text(i + bar_width, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

# Adicionar rótulos nas barras de tempo de resposta
for i, v in enumerate(media_tempo_resposta_por_mes):
    ax2.text(i + bar_width, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

# Adicionar rótulos nas barras de tempo de resposta
for i, v in enumerate(media_tempo_resposta_por_mes):
    ax2.text(i + bar_width, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

# Adicionar rótulos nas barras de tempo de resposta
for i, v in enumerate(media_tempo_resposta_por_mes):
    ax2.text(i + bar_width, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

# Unir as legendas de ambos os eixos
bars = bars1 + bars2
labels = [bar.get_label() for bar in bars]
ax1.legend(bars, labels, loc='upper left')

# Ajustar rótulos do eixo x para mostrar apenas os meses
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels([date.split()[0] for date in solicitacoes_por_mes.index], rotation=45, ha='right')

# Ajustar layout para evitar sobreposição de legendas
fig.tight_layout()

plt.title('Número de Pedidos de Informação (e-SIC) e Tempo Médio de Resposta por Mês (2022, 2023 e 2024)')
plt.xlabel('')

plt.show()

```

```python
# ANALISANDO AS SOLICITAÇÕES RESPONDIDAS POR AT

import pandas as pd
from collections import Counter

# 1. Filtrar para incluir apenas as linhas em que Área(s) responsável(eis) é igual a "AT"
dados_at = dados[dados['Área(s) responsável(eis)'] == 'AT']

# 2. Extrair os valores da coluna Tema(s)
temas_at = dados_at['Tema(s)']

# 3. Juntar todos os temas em uma única string
todos_temas = ' '.join(temas_at)

# 4. Dividir a string em palavras e contar a frequência
contagem_palavras = Counter(todos_temas.split())

# 5. Criar um DataFrame a partir do resultado
df_contagem_palavras = pd.DataFrame(list(contagem_palavras.items()), columns=['Palavra', 'Frequência'])

# 6. Ordenar o DataFrame pela coluna de frequência em ordem decrescente
df_contagem_palavras = df_contagem_palavras.sort_values(by='Frequência', ascending=False)

# Exibir a tabela
print(df_contagem_palavras)
```

```python
# NUVEM DE PALAVRAS DA VARÍAVEL "TEMA" DOS PEDIDOS RESPONDIDOS DIRETAMENTE POR AT

import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 1. Filtrar para incluir apenas as linhas em que Área(s) responsável(eis) é igual a "AT"
dados_at = dados[dados['Área(s) responsável(eis)'] == 'AT']

# 2. Extrair os valores da coluna Tema(s)
temas_at = dados_at['Tema(s)']

# 3. Juntar todos os temas em uma única string
todos_temas = ' '.join(temas_at)

# 4. Dividir a string em palavras e contar a frequência
contagem_palavras = Counter(todos_temas.split())

# 5. Exibir as palavras mais frequentes (opcional)
print(contagem_palavras.most_common(10))  # Exibe as 10 palavras mais frequentes

# 6. Criar uma nuvem de palavras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(contagem_palavras)

# Exibir a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
